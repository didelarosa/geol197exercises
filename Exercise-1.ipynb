```python
ice_cream_rating = range(0, 11)
sleeping_rating = range(0, 11)
```

```python
first_name = input("Enter your first name: ")
last_name = input ("Enter your last name: ")
```

```python
while True:
  try:
    ice_cream_rating = int(input("On a scale of 0 to 10, how would you rate your ice cream? "))
    break
  except ValueError:
      print("Please input a whole number from 0 to 10 only.")  
      continue
 
while True:
  try:
    sleeping_rating = int(input("On a scale of 0 to 10, how much do you enjoy your sleep? "))
    break
  except ValueError:
      print("Please input a whole number from 0 to 10 only.")  
      continue
```

```python
happiness_rating = (int(ice_cream_rating) + int(sleeping_rating))/2
print("Happiness rating:", happiness_rating)
percent_happiness = (happiness_rating/10)*100
```

```python
print(type(ice_cream_rating))
print(type(first_name))
print(type(happiness_rating))
```

The data types of the variables defined in this code are as expected. However, the data type for 'ice cream rating' initially appeared different (character string or 'str') when I hadn't validated the user input as integers yet.

```python
print("My name is", first_name, "and I give eating ice cream a score of", ice_cream_rating, "out of 10!")
print("I am", first_name, last_name, "and my sleeping enjoyment rating is", sleeping_rating, "/ 10!")
print("Based on the factors above, my happiness rating is", happiness_rating, "out of 10 or", percent_happiness, "%!") 
```
